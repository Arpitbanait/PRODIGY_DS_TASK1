```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/EV_sales_(King_county)[1].csv")
```

```python
df.head(5)
```

```python
df['Clean Alternative Fuel Vehicle Type'].unique()
```

```python
vehicle_counts = df['Clean Alternative Fuel Vehicle Type'].value_counts()


labels = vehicle_counts.index
values = vehicle_counts.values


plt.figure(figsize=(8, 4))
plt.bar(labels, values)
plt.title('Distribution of Clean Alternative Fuel Vehicle Types')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.xticks(rotation=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df['Make'].unique()
```

```python
electric_range = df['Electric Range'].dropna()


plt.figure(figsize=(5, 3))
plt.hist(electric_range, bins=30, color='orange', density=True, label="Histogram")
sns.kdeplot(electric_range, color='red', linewidth=1, label="KDE Line")
plt.title('Distribution of Electric Range')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
make_counts = df['Make'].value_counts()


labels = make_counts.index
values = make_counts.values


plt.figure(figsize=(8, 4))
plt.bar(labels, values,color="red")
plt.title('Distribution of vehicle manufacturing companies')
plt.xlabel('Manufacture')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
used_unused_counts = df['New or Used Vehicle'].value_counts()


labels = used_unused_counts.index
values = used_unused_counts.values


plt.figure(figsize=(5, 4))
plt.bar(labels, values,width = 0.3,color = "skyblue")
plt.title('Distribution of Clean Alternative Fuel Vehicle Types')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.xticks(rotation=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df['Transaction Year'].unique()
```

```python
transaction_year_counts = df['Transaction Year'].value_counts()


labels = transaction_year_counts.index
values = transaction_year_counts.values


plt.figure(figsize=(5, 4))
plt.bar(labels, values,width = 0.3,color = "lightgreen")
plt.title('Distribution of transaction over the years')
plt.xlabel('Years')
plt.ylabel('Count')
plt.xticks(rotation=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
